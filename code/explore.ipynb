## Data Exploration

Before cleaning our dataset, let's explore the features that our dataset contains, as well as visuals.

```python
# import required modules for this project

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load in required dataset 
df = pd.read_csv("../data/realestate.csv")

# print out the first 5 rows of this dataset 
df.head()
```

```python
# find out the datatypes of each column
df.info()
```

```python
# print out the summary statistics (mean, median, max, min, count) of all numeric columns
df.describe()
```

# Univariate Exploratory Analysis

Let's explore the distribution of single variables in this dataset.

```python
# histograms
df["price_per_unit_area"].plot.hist()
```

```python
# plot histogram of "house age" using sns here
df["house age"].plot.hist()
```

```python
# plot histogram of "num_convenience_stores" using sns here
df["num_convenience_stores"].plot.hist()
```

# Bivariate Exploratory Analysis

Let's explore the relationship of multiple variables in this dataset.

```python
# scatter plot of house age & price of unit area
sns.scatterplot(data=df, x="house age", y="price_per_unit_area")
plt.show()
```

```python
# Tplot scatter plot of "num_convenience_stores" and "price_per_unit_area"

sns.scatterplot(data=df, x="num_convenience_stores", y="price_per_unit_area")
plt.show()
```
